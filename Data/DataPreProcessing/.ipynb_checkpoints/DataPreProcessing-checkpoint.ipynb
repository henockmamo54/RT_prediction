```python
import os
import utils as ut
import pandas as pd
import statistics
import numpy as np
import matplotlib.pyplot as plt
from itertools import combinations
from sklearn.model_selection import train_test_split
from statsmodels.nonparametric.smoothers_lowess import lowess
import warnings

warnings.filterwarnings('ignore')
```

## Extract retention time outputs from d2ome quantification files

```python
filepath = r'..\d2ome_output'
organs = ['liver', 'kidney', 'heart', 'muscle']
```

```python
data={
'liver_rt':ut.get_rt_info(f'{filepath}\\liverpool_liver'),
'kidney_rt':ut.get_rt_info(f'{filepath}\\liverpool_kidney'),
'heart_rt':ut.get_rt_info(f'{filepath}\\liverpool_heart') , 
'muscle_rt':ut.get_rt_info(f'{filepath}\\liverpool_muscle')}
```

# Retention time alignment between different experiments

```python
# Initialize all_data with a copy of the 'liver_rt' DataFrame.
# This DataFrame will be progressively augmented with data from other tissues.
all_data = data['liver_rt'].copy()

# Define a list of tuples, where each tuple represents a pair of tissue data keys to be compared and merged.
# The first element in the tuple is the reference tissue (e.g., 'liver_rt'),
# and the second is the tissue to be aligned/merged with the reference.
data_key_combinations = [('liver_rt', 'kidney_rt'), ('liver_rt', 'heart_rt'), ('liver_rt', 'muscle_rt')]

# Iterate through each combination of tissue data.
for index in range(len(data_key_combinations)):

    # Merge two DataFrames based on common 'Protein', 'Peptide', and 'Charge' columns.
    # This step aligns retention times (rt) from two different tissues for the same identified peptide.
    # data_key_combinations[index][1] is the 'x' dataset (e.g., 'kidney_rt')
    # data_key_combinations[index][0] is the 'y' dataset (e.g., 'liver_rt')
    merged_data = pd.merge(data[data_key_combinations[index][1]],
                                data[data_key_combinations[index][0]], on=['Protein', 'Peptide', 'Charge'], how='inner')

    # Extract the retention times from the merged data to be used for model fitting.
    # 'rt_x' corresponds to the retention time from the second tissue in the current combination (e.g., kidney).
    # 'rt_y' corresponds to the retention time from the first tissue (reference, e.g., liver).
    x, y = np.array(merged_data['rt_x']), np.array(merged_data['rt_y'])

    # Fit a polynomial model to the retention times (x vs y) and plot the fit.
    # 'ut.plot_polynomial_fit' is assumed to be a utility function that performs this.
    # Labels for the plot are dynamically generated based on the current tissue combination.
    model = ut.plot_polynomial_fit(x, y, verbose=True,
                                 x_label=data_key_combinations[index][1] + "(minutes)",
                                 y_label=data_key_combinations[index][0] + "(minutes)")

    # --- Predict retention times for data imputation ---
    # Create a temporary copy of the second tissue's data (e.g., 'kidney_rt').
    temp_data = data[data_key_combinations[index][1]].copy()
    # Use the fitted polynomial model to predict retention times ('rt') for this temporary dataset.
    # The original 'rt' column of temp_data is used as input for prediction.
    temp_data['rt'] = model.predict(np.array(temp_data['rt']).reshape(-1, 1))

    # Merge the temporary data (with predicted 'rt') with the 'all_data' DataFrame.
    # This is a left merge to identify entries in 'temp_data' that are not already in 'all_data'.
    temp_data = pd.merge(temp_data, all_data, on=['Protein', 'Peptide', 'Charge'], how='left')
    # Filter 'temp_data' to keep only rows where 'rt_y' is NaN.
    # 'rt_y' comes from 'all_data'; NaN here means the peptide-charge state was in the current tissue (e.g., kidney)
    # but not yet in 'all_data' (which started with liver and gets augmented).
    temp_data = temp_data[temp_data['rt_y'].isna()]

    # Select and rename columns to match the schema of 'all_data'.
    # We are interested in 'Protein', 'Peptide', 'Charge', 'm/z(Sequence)', 'IonScore', and the newly predicted 'rt'.
    # Suffixes '_x' indicate columns originating from the left DataFrame in the previous merge (temp_data).
    temp_data = temp_data[['Protein', 'Peptide', 'Charge', 'm/z(Sequence)_x', 'IonScore_x', 'rt_x']].rename(columns={
        'm/z(Sequence)_x': 'm/z(Sequence)', 'IonScore_x': 'IonScore',
        'rt_x': 'rt'})

    # Concatenate the filtered and processed 'temp_data' (containing new entries with predicted rt)
    # to the main 'all_data' DataFrame.
    all_data = pd.concat([all_data, temp_data], ignore_index=True)

    # Display the plot generated by 'ut.plot_polynomial_fit'.
    plt.show()
```

## Export the preprocessed dataset

```python
all_data=all_data.groupby(['Protein',	'Peptide',	'Charge',	'm/z(Sequence)'])['rt'].median('aligned_rt') .reset_index()
all_data.to_csv(f'liverpool_all_data.csv', index=False)
```
